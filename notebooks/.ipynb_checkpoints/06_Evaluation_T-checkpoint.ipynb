```python
import pandas as pd
import numpy as np
import pickle
import warnings
warnings.filterwarnings("ignore")
```

```python
ratings = pd.read_csv("../data/raw/ratings.csv")
books = pd.read_csv("../data/raw/books.csv")

ratings = ratings[ratings['rating'] >= 4]

data = ratings.merge(
    books[['book_id', 'title']],
    on='book_id'
)
```

```python
with open("hybrid_model.pkl", "rb") as f:
    model = pickle.load(f)

rules = model["rules"]
user_similarity_df = model["user_similarity"]
user_item_matrix = model["user_item_matrix"]

print("Hybrid model loaded successfully.")
```

```python
def collaborative_recommend(user_id, top_n=5):
    
    similar_users = user_similarity_df[user_id].sort_values(
        ascending=False
    )[1:21]
    
    weighted_sum = np.dot(
        similar_users.values,
        user_item_matrix.loc[similar_users.index]
    )
    
    scores = pd.Series(
        weighted_sum,
        index=user_item_matrix.columns
    )
    
    already_read = user_item_matrix.loc[user_id]
    scores = scores[already_read == 0]
    
    return scores.sort_values(ascending=False).head(top_n).index.tolist()


def association_recommend(book_name, top_n=5):
    
    filtered = rules[
        rules['antecedents'].map(lambda x: book_name in x)
    ]
    
    if filtered.empty:
        return []
    
    filtered = filtered.sort_values(by='lift', ascending=False)
    
    recommendations = []
    
    for cons in filtered['consequents']:
        recommendations.extend(list(cons))
    
    recommendations = list(dict.fromkeys(recommendations))
    
    return recommendations[:top_n]


def hybrid_recommend(user_id=None, book_name=None, top_n=5):

    scores = {}

    if user_id is not None and user_id in user_item_matrix.index:
        cf_recs = collaborative_recommend(user_id, top_n=30)
        for i, book in enumerate(cf_recs):
            scores[book] = scores.get(book, 0) + (1 / (i + 1))

    if book_name is not None:
        ar_recs = association_recommend(book_name, top_n=30)
        for i, book in enumerate(ar_recs):
            scores[book] = scores.get(book, 0) + (1 / (i + 1))

    if not scores:
        return data['title'].value_counts().head(top_n).index.tolist()

    sorted_books = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    
    return [book for book, _ in sorted_books[:top_n]]
```

```python
def recall_at_k(user_id, k=15):
    
    user_books = data[data['user_id'] == user_id]['title'].tolist()
    
    if len(user_books) < 3:
        return None
    
    test_books = set(user_books[-2:])
    train_books = user_books[:-2]
    
    recs = hybrid_recommend(
        user_id=user_id,
        book_name=train_books[-1],
        top_n=k
    )
    
    return len(set(recs) & test_books) / len(test_books)
```

```python
def precision_at_k(user_id, k=10):
    
    user_books = data[data['user_id'] == user_id]['title'].tolist()
    
    if len(user_books) < 3:
        return None
    
    test_books = set(user_books[-2:])
    train_books = user_books[:-2]
    
    recs = hybrid_recommend(
        user_id=user_id,
        book_name=train_books[-1],
        top_n=k
    )
    
    return len(set(recs) & test_books) / k
```

```python
users = user_item_matrix.index[:200]

recall_scores = []
precision_scores = []

for u in users:
    r = recall_at_k(u, k=15)
    p = precision_at_k(u, k=15)
    
    if r is not None:
        recall_scores.append(r)
    if p is not None:
        precision_scores.append(p)

print("Average Recall@15:", np.mean(recall_scores))
print("Average Precision@15:", np.mean(precision_scores))
```

```python
def coverage(k=20):
    recommended = set()
    
    for u in users:
        recs = hybrid_recommend(user_id=u, top_n=k)
        recommended.update(recs)
    
    return len(recommended) / len(user_item_matrix.columns)

print("Catalog Coverage:", coverage())
```

```python
def diversity(user_id, k=10):
    
    recs = hybrid_recommend(user_id=user_id, top_n=k)
    
    if len(recs) < 2:
        return 0
    
    pairs = 0
    dissimilarity = 0
    
    for i in range(len(recs)):
        for j in range(i+1, len(recs)):
            pairs += 1
            dissimilarity += 1  # assume different books
    
    return dissimilarity / pairs

print("Diversity Example:", diversity(users[0]))
```

```python
print("="*40) 
print("FINAL HYBRID MODEL EVALUATION")
print("="*40)
print(f"Users Evaluated     : {len(users)}")
print(f"Recall@15           : {np.mean(recall_scores):.4f}")
print(f"Precision@15        : {np.mean(precision_scores):.4f}")
print(f"Catalog Coverage    : {coverage():.4f}")
print(f"Diversity (Sample)  : {diversity(users[0]):.4f}")
print("="*40)
```

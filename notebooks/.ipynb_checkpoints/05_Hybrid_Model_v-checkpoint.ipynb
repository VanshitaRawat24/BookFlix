```python
import pandas as pd
import numpy as np
import pickle
from sklearn.metrics.pairwise import cosine_similarity
import warnings
warnings.filterwarnings("ignore")
```

```python
ratings = pd.read_csv("../data/raw/ratings.csv")
books = pd.read_csv("../data/raw/books.csv")
# Keep positive ratings
ratings = ratings[ratings['rating'] >= 4]

# Merge titles
data = ratings.merge(
    books[['book_id', 'title']],
    on='book_id'
)

print("Data shape:", data.shape)
```

```python
user_item_matrix = data.pivot_table(
    index='user_id',
    columns='title',
    values='rating'
).fillna(0)

print("User-Item Matrix Shape:", user_item_matrix.shape)
```

```python
user_similarity = cosine_similarity(user_item_matrix)

user_similarity_df = pd.DataFrame(
    user_similarity,
    index=user_item_matrix.index,
    columns=user_item_matrix.index
)
```

```python
def collaborative_recommend(user_id, top_n=5):
    
    similar_users = user_similarity_df[user_id].sort_values(
        ascending=False
    )[1:11]
    
    weighted_sum = np.dot(
        similar_users.values,
        user_item_matrix.loc[similar_users.index]
    )
    
    scores = pd.Series(
        weighted_sum,
        index=user_item_matrix.columns
    )
    
    already_read = user_item_matrix.loc[user_id]
    scores = scores[already_read == 0]
    
    return scores.sort_values(ascending=False).head(top_n).index.tolist()
```

```python
def association_recommend(book_name, top_n=5):
    
    filtered = rules[
        rules['antecedents'].map(lambda x: book_name in x)
    ]
    
    if filtered.empty:
        return []
    
    filtered = filtered.sort_values(
        by='lift', ascending=False
    )
    
    recommendations = []
    
    for cons in filtered['consequents']:
        recommendations.extend(list(cons))
    
    recommendations = list(dict.fromkeys(recommendations))
    
    return recommendations[:top_n]
```

```python
def popularity_recommend(top_n=5):
    return data['title'].value_counts().head(top_n).index.tolist()
```

```python
def hybrid_recommend(user_id=None, book_name=None, top_n=5):

    scores = {}

    # Collaborative contribution
    if user_id is not None and user_id in user_item_matrix.index:
        cf_recs = collaborative_recommend(user_id, top_n=20)
        for i, book in enumerate(cf_recs):
            scores[book] = scores.get(book, 0) + (1 / (i + 1))

    # Association contribution
    if book_name is not None:
        ar_recs = association_recommend(book_name, top_n=20)
        for i, book in enumerate(ar_recs):
            scores[book] = scores.get(book, 0) + (1 / (i + 1))

    # If nothing found → fallback
    if not scores:
        return popularity_recommend(top_n)

    sorted_books = sorted(
        scores.items(),
        key=lambda x: x[1],
        reverse=True
    )

    return [book for book, _ in sorted_books[:top_n]]
```

```python
hybrid_recommend(user_id=1, book_name="The Hobbit", top_n=5)
```

```python
def recall_at_k(user_id, k=10):
    
    user_books = data[data['user_id'] == user_id]['title'].tolist()
    
    if len(user_books) < 3:
        return None
    
    test_books = set(user_books[-2:])   # hide last 2
    train_books = user_books[:-2]
    
    recs = hybrid_recommend(
        user_id=user_id,
        book_name=train_books[-1],
        top_n=k
    )
    
    return len(set(recs) & test_books) / len(test_books)
```

```python
users = user_item_matrix.index[:100]

scores = []
for u in users:
    r = recall_at_k(u, k=15)
    if r is not None:
        scores.append(r)

print("Average Recall@15:", sum(scores)/len(scores))
```

```python
def coverage(k=20):
    recommended = set()
    
    for u in user_item_matrix.index[:100]:
        recs = hybrid_recommend(user_id=u, top_n=k)
        recommended.update(recs)
    
    return len(recommended) / len(user_item_matrix.columns)

print("Hybrid Coverage:", coverage())
```

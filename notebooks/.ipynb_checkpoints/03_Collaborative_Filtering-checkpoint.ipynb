```python
import pandas as pd
import numpy as np

ratings = pd.read_csv("../data/processed/ratings_clean.csv")
books = pd.read_csv("../data/processed/books_enhanced.csv")
books = books.set_index("book_id")
```

```python
from sklearn.model_selection import train_test_split

ratings_train = []
ratings_test = []

for user in ratings['user_id'].unique():
    
    user_data = ratings[ratings['user_id'] == user]
    
    if len(user_data) > 5:
        train, test = train_test_split(
            user_data,
            test_size=0.2,
            random_state=42
        )
        
        ratings_train.append(train)
        ratings_test.append(test)

ratings_train = pd.concat(ratings_train)
ratings_test = pd.concat(ratings_test)
```

```python
user_item_matrix = ratings_train.pivot_table(
    index="user_id",
    columns="book_id",
    values="rating"
).fillna(0)
```

```python
from sklearn.decomposition import TruncatedSVD

svd = TruncatedSVD(n_components=30, random_state=42)

latent_matrix = svd.fit_transform(user_item_matrix)

reconstructed_matrix = np.dot(
    latent_matrix,
    svd.components_
)

# Clip predictions to rating range
reconstructed_matrix = np.clip(reconstructed_matrix, 1, 5)

reconstructed_df = pd.DataFrame(
    reconstructed_matrix,
    index=user_item_matrix.index,
    columns=user_item_matrix.columns
)

print("SVD trained successfully")
```

```python
def recommend_books_svd(user_id, top_n=10):
    
    if user_id not in reconstructed_df.index:
        return "User not found"
    
    user_ratings = reconstructed_df.loc[user_id]
    
    already_rated = user_item_matrix.loc[user_id] > 0
    
    recommendations = user_ratings[~already_rated] \
        .sort_values(ascending=False) \
        .head(top_n).index
    
    return recommendations
```

```python
def precision_at_k_test(user_id, k=10):
    
    test_books = ratings_test[
        ratings_test['user_id'] == user_id
    ]['book_id']
    
    recommended = recommend_books_svd(user_id, top_n=k)
    
    if isinstance(recommended, str):
        return None
    
    return len(set(test_books) & set(recommended)) / k
```

```python
precisions = []

for user in ratings_test['user_id'].unique():
    p = precision_at_k_test(user,10)
    if p is not None:
        precisions.append(p)

print("Average Precision@10:", round(sum(precisions)/len(precisions),3))
```

```python
def recall_at_k_test(user_id, k=10):
    
    test_books = ratings_test[
        ratings_test['user_id'] == user_id
    ]['book_id']
    
    if len(test_books) == 0:
        return None
    
    recommended = recommend_books_svd(user_id, top_n=k)
    
    if isinstance(recommended, str):
        return None
    
    return len(set(test_books) & set(recommended)) / len(test_books)
```

```python
def hit_rate(user_id, k=10):
    
    test_books = ratings_test[
        ratings_test['user_id'] == user_id
    ]['book_id']
    
    recommended = recommend_books_svd(user_id, top_n=k)
    
    if isinstance(recommended, str):
        return None
    
    return int(len(set(test_books) & set(recommended)) > 0)
```

```python
def debug_hit_position(user_id, k=20):
    
    test_books = set(
        ratings_test[ratings_test['user_id'] == user_id]['book_id']
    )
    
    recommended = recommend_books_svd(user_id, top_n=k)
    
    if isinstance(recommended, str):
        print("User not found")
        return
    
    for rank, book_id in enumerate(recommended, start=1):
        
        if book_id in test_books:
            print(f"Hit at rank: {rank}")
            return
    
    print(f"No hit in top {k}")
```
